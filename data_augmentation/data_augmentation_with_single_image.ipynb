```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
```

```python
image1 = image.load_img('train/dog_image.jpg',target_size=(200,200))
```

```python
img_array = image.img_to_array(image1)
print(img_array.shape)
```

```python
print("Top-left pixel:", img_array[0, 0])      # [R, G, B] values
print("Center pixel:", img_array[128, 128])
```

```python
import numpy as np
np.set_printoptions(threshold=np.inf)  # disable truncation
print(img_array)

```

```python
type(image1)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)
```

```python
img = image.img_to_array(image1)
```

```python
type(img)
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
input_batch.shape
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size = 1, save_to_dir= 'aug'):
    i = i + 1

    if i ==10:
      break
```
